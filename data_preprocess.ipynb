```python
def eeg_data_preprocess():
    root = "./data/data_preprocessed_python/"
    des = "./data/DEAP/bio/"
    labels = pd.read_csv("./data/DEAP/labels/participant_ratings.csv")
    labels.head()
    for file in os.listdir(root):
        subject = file.split(".")[0]
        subject_id = int(subject[1:])
        try :
            os.mkdir(des + "s" + str(subject_id))
        except Exception as e:
            print(des + "s" + str(subject_id), "already created")
 
        f = open(root + file, "rb")
        d = cPickle.load(f, encoding = "latin")

        data = d["data"]
        for experiment in range(40):

            trial = labels[(labels["Participant_id"] == subject_id) & (labels["Experiment_id"] == experiment + 1)]["Trial"].iloc[0]
            l = []
            for i in range(3):
                l.append(data[experiment][:, i * 128: (i + 1)* 128])
            baseline_mean = sum(l)/3
            
            for i in range(60):
                data_seg = data[experiment][:, 384 + i * 128: 384  + (i + 1) * 128]
                data_seg_removed = data_seg - baseline_mean
                print(data_seg_removed.shape)
                # np.save(f'{des}s{subject_id}/{subject_id}_{trial}_{i + 1}.npy', data_seg_removed)
        
eeg_data_preprocess()
```
